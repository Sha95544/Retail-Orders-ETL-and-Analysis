```python
# extracting the orders.csv file from the zip file:
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() # extracting the file to the current directory
zip_ref.close() # closing the file
```

```python
import pandas as pd
df = pd.read_csv('orders.csv')
df.head(10)
```

```python
# finding unique values in ship mode column:
df['Ship Mode'].unique()
```

```python
# Treat "Not Available", "unknown" and nan in Ship Mode column as null values
# For that we will read the csv file again specifying the valuyes to be treated as null values:
df = pd.read_csv("orders.csv",na_values=["Not Available","unknown"])
df["Ship Mode"].unique()
```

```python
# renaming gthe columns i.e making them lowwer case and replacing space with _
# The approach below is not eficient since we will need to speicfy all th 15 columns for the function
# df.rename(columns={"Order Id":"order_id"})

df.columns = df.columns.str.lower()
# df.columns # (converted to lower case)
df.columns = df.columns.str.replace(" ","_")
# df.columns
df.head()
```

```python
# creating new columns for discount, sale price and profit
# discount:
df["discount"] = df["list_price"]*df["discount_percent"]*0.01
df.head()
```

```python
# adding sale price and profit columns
# df["sale_price"] = df["list_price"]-df["discount"]
df["profit"] = df["sale_price"] - df["cost_price"]
```

```python
df.head()
```

```python
# checking column types:
df.dtypes
```

```python
# currently order_date is stored as a string object so it will be converted to a datetime object:
df["order_date"] = pd.to_datetime(df["order_date"])
```

```python
# verifying the date type conversion:
df.dtypes
```

```python
# Dropping the unnecessary columns such as cost_price, list_price and discount_percent
df.drop(columns=["list_price", "cost_price", "discount_percent"], inplace=True)
```
